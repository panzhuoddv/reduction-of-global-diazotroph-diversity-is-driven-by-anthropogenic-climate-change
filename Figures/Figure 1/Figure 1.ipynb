```python
# Common imports
import numpy as np
import os
# To plot pretty figures
# %matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
# 轴线宽度
mpl.rc('xtick', labelsize=12)
# X刻度值宽度
mpl.rc('ytick', labelsize=12)
# Y刻度值宽度
# rc 修改默认的属性，包括窗体大小、每英寸的点数、线条宽度、颜色、样式、坐标轴、坐标和网络属性、文本、字体等。
# Where to save the figures
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "heatmap"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
# os.path.join组装路径 以上组装效果为 ./images/peptide_tm  其中./代表当前目录 可省略
os.makedirs(IMAGES_PATH, exist_ok=True)
# os.makedirs用于创建递归目录  相对路径和绝对路径均可  第一个参数是路径   
# 若目录存在则不再创建 exist_ok=True表明如果目录存在不报错 默认为False 即目录存在则报错
def save_fig(fig_id, tight_layout=True, fig_extension="pdf", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
import warnings
warnings.filterwarnings(action="ignore", message="^internal gelsd")
import seaborn as sns
```

```python

fig, ax = plt.subplots(figsize=(12, 7),
                       subplot_kw=dict(aspect="equal"),
                       dpi=300)

data = [14383,1723]
categories = ['Genome without nif gene','Genome with nif gene']
explode = [0, 0.1]


def func(pct, allvals):
    absolute = int(pct / 100. * np.sum(allvals))
    return "{:.1f}% ({:d} )".format(pct, absolute)

colors = ['#C0C0C0','red']
wedges, texts, autotexts = ax.pie(data,
                                  autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"),
                                  colors=colors,
                                  startangle=15,
                                  explode=explode)

# Decoration
ax.legend(wedges,
          categories,
          title="Vehicle Class",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=16, weight=700)
ax.set_title("Class of Vehicles: Pie Chart")
save_fig('all_pie2')
```
